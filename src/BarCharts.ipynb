````python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
````

## Dataset
We work with a dataset from the US Department of Transportation that tracks flight delays.

Each entry shows the average arrival delay (in minutes) for a different airline and month (all in year 2015). Negative entries denote flights that (on average) tended to arrive early.

````python
# Path of the file to read
flight_filepath = "../data/flight_delays.csv"

# Read file into a dataframe
flight_data = pd.read_csv(flight_filepath, index_col="Month")

flight_data
````

## Bar chart

Let's create a bar chart showing the average arrival delay for airline code: NK, by month.

The code to create the bar chart has three main components:
* ```sns.barplot```- tells the notebook that we want to create a bar chart.
* ```x=flight_data.index```- this determines what to use on the horizontal axis. In this case, we have selected the column that indexes the rows.
* ```y=flight_data['NK']```- this sets the column in the data that will be used to determine the height of each bar. In this case, we select the 'NK' column.

> Note:
    We must always select the indexing column with ```flight_data.index```, and it is not 
    possible to use ```flight_data[\'Month\']``` which will return an error. This is 
    because when we loaded the dataset, the "Month" column was used to index the rows.


````python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Average Arrival Delay for NK Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
````

## Heatmap

We create a heatmap to quickly visualize patterns in ```flight_data```. Each cell is color-coded accoding to its corresponding value.

The relevant code to create the heatmap has three main components:
* ```sns.heatmap```- tells the notebook that we want to create a heatmap.
* ```data=flight_data``` - tells the notebook to use all of the entries in ```flight_data``` to create the heatmap.
* ```annot=True```- this ensures that the values for each cell appear on the chart. (Leaving this removes the numbers from each of the cells)

````python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True)

# Add label for horizontal axis
plt.xlabel("Airline")
````

The months towards the end of the year (especially months 9-11) appear relatively dark for all airlines. This suggests that airlines are better (on average) at keeping schedule during these months!

We want to develop a game that would work best on all platforms. To do so, we use the IGN Game Reviews dataset and analyse it to design our upcoming game.

````python
# Path of the file to read
ign_filepath = "../data/ign_scores.csv"

# Fill in the line below to read the file into a variable ign_data
ign_data = pd.read_csv(ign_filepath, index_col="Platform")

# Print the data
ign_data
````

The dataset shows the average score, by platform and genre.

We now create a bar chart that shows the average score for racing games, for each platform.

````python
# Set the width and height of the figure
plt.figure(figsize=(26,10))

# Add title
plt.title("Average Score for Racing Games, by Platform")

# Bar chart showing average score for racing games
sns.barplot(x=ign_data.index, y=ign_data['Racing'])

# Add label for vertical axis
plt.ylabel("Arrival Score (by Platform)")
````

Based on the data, we should not expect a racing game for the Wii platform to receive a high rating. In fact, on average, racing game for Wii score lower than any other platform.
Xbox One seems to be the best alternative, since it has the highest average ratings.

We now creat a heatmap of average score by genre and platform

````python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("Average Score by Genre and Platform")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=ign_data, annot=True)

# Add label for horizontal axis
plt.xlabel("Genre")
````

Simulation games for PlayStation 4 receive the highest average ratings (9.2).
Shooting and Fighting games for Game Boy Color reveive the lowest average rankings (4.5).
